```python
import os
import pathlib

import numpy as np

from PIL import Image
```

```python
### On récupère les images du dataset
```

```python
training_path = pathlib.Path("kaggle/cats_and_dogs/dataset/training_set")
testing_path = pathlib.Path("kaggle/cats_and_dogs/dataset/test_set")

names=os.listdir(training_path)

for name in names:
    image_count = len(list(training_path.glob('*/*.jpg')))
print("classes : ", names)
print("images : ", image_count)
```

```python
dataset = []
count = 0
for name in names:
    for filename in os.listdir(os.path.join(training_path,name)):
        count += 1
        print("\r" + str(count) + " / " + str(image_count), end="")
        image = Image.open(os.path.join(training_path,name,filename))
        image.load()
        image = np.asarray(image, dtype="float32" )
        dataset.append((image, name))

print(dataset)
```

```python
import random
print(random.sample(dataset, 10))
random.shuffle(dataset)
```
